```python
from transformers import AutoProcessor, AutoModelForImageTextToText
import torch

model_path = "HuggingFaceTB/SmolVLM2-2.2B-Instruct"
processor = AutoProcessor.from_pretrained(model_path)
model = AutoModelForImageTextToText.from_pretrained(
    model_path,
    torch_dtype=torch.bfloat16,
    _attn_implementation="flash_attention_2"
).to("cuda")
```

```python
from glob import glob
import sys

question_list= [
    "Can you describe this image?",
     "Explain the purpose of image",
     "Can you comment on size and kind of this image",
    "Create a interesting story about this image",
    "Elaborate the setting of the image"

]

data_list = []
for file in glob('*.png'):
    for question in question_list:
        messages = [
            {
                "role": "user",
                "content": [
                    {"type": "image", "url": "/content/" + file},
                    {"type": "text", "text": question},
                 ]
            },
         ]

        inputs = processor.apply_chat_template(
            messages,
            add_generation_prompt=True,
            tokenize=True,
            return_dict=True,
            return_tensors="pt",
        ).to(model.device, dtype=torch.bfloat16)

        generated_ids = model.generate(**inputs, do_sample=False, max_new_tokens=64)
        generated_texts = processor.batch_decode(
            generated_ids,
            skip_special_tokens=True,
        )

        answer = generated_texts[0]
        answer = answer[answer.index("Assistant: ") + len("Assistant:"):]
        print(f"image: {file}, question: {question}, answer: {answer}")
        data = {}
        data["image"] = file
        data["question"] = question
        data["answer"] = answer
        data_list.append(data)
```

```python
import pandas as pd
df = pd.DataFrame(data_list)
df.to_csv('image_text_dataset.csv', index = False)
```
